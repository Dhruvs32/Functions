Q.1. What is the difference between a function and a method in Python?
  - n Python, functions and methods are both blocks of reusable code that perform a specific task, but they differ mainly in how they are called and what they are associated with.

1. Function
A function is a block of code that is defined using the def keyword or created using lambda.

It is not tied to any object or class (though it can be used in object-oriented contexts).

Can be called independently using its name.

Example:

python
Copy
Edit
def greet(name):
    return f"Hello, {name}!"

greet("Alice")  # Output: Hello, Alice!
2. Method
A method is a function that is associated with an object (i.e., it’s defined within a class).

It is called on an object (an instance of a class) and can access and modify the object’s state using self.

Methods are functions that "belong to" objects.

Example:

python
Copy
Edit
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

greeter = Greeter()
greeter.greet("Alice")  # Output: Hello, Alice!
Key Differences:
Feature	Function	Method
Belongs to	Not bound to any object	Bound to a class or object
Call syntax	function_name(args)	object.method_name(args)
First argument	No automatic first argument	First argument is usually self
Context	Can be global or inside a class	Must be inside a class

So in summary: All methods are functions, but not all functions are methods.

Q.5. What are iterators in Python and how do they differ from iterables?
   - In Python, iterators and iterables are key concepts when working with loops, especially for loops. They sound similar but serve different roles.

✅ 1. Iterable
An iterable is any Python object capable of returning its elements one at a time, allowing it to be looped over (e.g., in a for loop).

Common Examples:
Lists: [1, 2, 3]

Tuples: (4, 5)

Strings: "hello"

Dictionaries: {"a": 1, "b": 2}

Sets: {1, 2, 3}

Key Point:
An iterable has a __iter__() method that returns an iterator.

python
Copy
Edit
my_list = [1, 2, 3]
it = iter(my_list)  # Get an iterator from the iterable
✅ 2. Iterator
An iterator is an object that keeps state and produces the next value when you call next() on it.

Key Features:
Implements both __iter__() and __next__() methods.

Remembers where it is during iteration.

Raises a StopIteration exception when there are no more items.

python
Copy
Edit
it = iter([1, 2, 3])  # it is now an iterator
print(next(it))  # Output: 1
print(next(it))  # Output: 2
print(next(it))  # Output: 3
# print(next(it)) would raise StopIteration
🔄 Key Differences: Iterable vs Iterator
Feature	Iterable	Iterator
What it does	Can be looped over	Returns items one at a time
Methods	Has __iter__()	Has both __iter__() and __next__()
Returns	Returns an iterator	Returns next item on demand
Examples	List, Tuple, Set, String	Output of iter()
Can use next()?	❌ No	✅ Yes

✅ Custom Iterator Example
python
Copy
Edit
class Counter:
    def __init__(self, low, high):
        self.current = low
        self.high = high

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.high:
            raise StopIteration
        else:
            self.current += 1
            return self.current - 1

for num in Counter(1, 3):
    print(num)
# Output: 1, 2, 3

Q.6.  Explain the concept of generators in Python and how they are defined.
  - Generators are a special type of iterators that allow you to yield values one at a time, only when needed — making them memory-efficient and ideal for working with large data sets or infinite sequences.

✅ Key Characteristics of Generators
Defined like a normal function but use the yield keyword instead of return.

Automatically create an iterator without requiring __iter__() and __next__() methods.

Keep their state between calls, so they resume where they left off.

🛠 How to Define a Generator
1. Using a generator function
python
Copy
Edit
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

gen = count_up_to(3)
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
# next(gen) → StopIteration
2. Using generator expressions (like list comprehensions)
python
Copy
Edit
gen = (x*x for x in range(5))
print(next(gen))  # Output: 0
⚙️ How Generators Work
Each time yield is hit:

The function's state is saved.

The value after yield is returned.

On the next call (e.g., via next()), execution resumes right after the last yield.

🔁 Difference Between return and yield
Feature	return	yield
Terminates?	Ends the function immediately	Pauses the function, saving state
Use case	Single result	Sequence of values (lazy)
Memory	Stores all values at once	Produces values one at a time

✅ Why Use Generators?
Efficient for large datasets (e.g., streaming files, logs).

Natural fit for pipelines and lazy evaluation.

Easier to write and read than manual iterator classes.



Q.7. What are the advantages of using generators over regular functions?
  - Generators offer several key benefits, especially when working with large data, infinite sequences, or performance-critical code. Here's why you might choose a generator instead of a regular function:

🚀 1. Memory Efficiency (Lazy Evaluation)
Generators produce one item at a time, on demand — instead of generating and storing the entire result in memory.

Example:
python
Copy
Edit
def count_up_to(n):
    for i in range(n):
        yield i

# Efficient even with huge ranges
for num in count_up_to(1_000_000):
    pass
🔹 Advantage: Doesn't create a large list in memory, reducing RAM usage.

⚡ 2. Faster Startup Time
Since generators yield values one at a time, you don't wait for the whole dataset to be prepared before starting to process it.

python
Copy
Edit
def data_stream():
    yield "record 1"
    yield "record 2"
🔹 Advantage: You can start working on data as soon as the first value is available.

🔁 3. Infinite Sequences Support
Generators can model infinite data streams (e.g., sensor readings, live logs) without ever running out of memory.

python
Copy
Edit
def infinite_counter():
    i = 0
    while True:
        yield i
        i += 1
✅ 4. Simpler Code Than Manual Iterators
Writing a generator with yield is much simpler than implementing an iterator class with __iter__() and __next__().

Iterator Class (Verbose):

python
Copy
Edit
class Counter:
    def __init__(self):
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        self.i += 1
        return self.i
Equivalent Generator (Concise):

python
Copy
Edit
def counter():
    i = 0
    while True:
        i += 1
        yield i
🔄 5. Natural Fit for Pipelines and Streams
Generators allow you to build data pipelines, where each step processes data lazily and efficiently.

python
Copy
Edit
lines = (line for line in open("file.txt"))
words = (line.split() for line in lines)
📋 Summary: Advantages of Generators
Advantage	Benefit
Memory Efficiency	No need to store large data sets in memory
Lazy Evaluation	Values are computed only when needed
Simpler Code	Less boilerplate than iterator classes
Infinite Sequences	Can handle unbounded data
Stream Processing	Ideal for reading files, streams, or pipelines

Q.9. Explain the purpose and usage of the `map()` function in Python.

✅ Purpose of map()

Helps avoid explicit loops, making code cleaner and often more efficient.

🧠 Syntax
python
Copy
Edit

function: A function that takes one argument (can also be a lambda).

iterable: A sequence like a list, tuple, or string.

You can convert the result to a list or another iterable:

python
Copy
Edit
list(map(function, iterable))
🧪 Example: Basic Usage
python
Copy
Edit
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
squared = map(square, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
🔸 Example with lambda Function
python
Copy
Edit

squared = map(lambda x: x * x, numbers)

You can pass multiple iterables to map() if the function accepts multiple arguments:

python
Copy
Edit

✅ When to Use map()
Scenario	Use map() Because...
You need to transform data	Applies a function to every item in a sequence
You want cleaner code	Avoids writing verbose for loops
You're dealing with large data	Returns an iterator → memory-efficient
You're using functional patterns	Works well with lambda, filter(), reduce()

⚠️ Things to Remember
map() is lazy — it returns an iterator, not a list. Use list(), tuple(), etc., to consume it.

If you pass more than one iterable, they must be of the same length (shorter one determines length).

Q.10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - 🔄 1. map()
✅ Purpose:
Transforms each item in an iterable using a specified function.

🧪 Example:
python
Copy
Edit

✅ Use case:
Modify or compute values from each element.

🔍 2. filter()
✅ Purpose:
Filters items from an iterable based on a condition (a function that returns True or False).

🧪 Example:
python
Copy
Edit

print(list(evens))  # [2, 4]
✅ Use case:
Keep only elements that match a condition.

➕ 3. reduce() (from functools module)
✅ Purpose:
Reduces a sequence to a single value by applying a function cumulatively.

🧪 Example:
python
Copy
Edit
from functools import reduce

nums = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, nums)
print(total)  # 10
✅ Use case:
Combine all items into one (sum, product, max, etc.)

🔁 Comparison Table
Feature	map()	filter()	reduce()
Purpose	Transform each element	Select elements based on condition	Combine elements into one value
Returns	map object (iterator)	filter object (iterator)	Single value
Function type	f(x)	f(x) → bool	f(x, y)
Common use	Squaring, formatting, conversions	Removing unwanted items	Summing, multiplying, aggregating
Needs import?	No	No	✅ Yes, from functools

🧠 Example Using All Three
python
Copy
Edit
from functools import reduce

nums = [1, 2, 3, 4, 5]

# Square numbers
squares = map(lambda x: x**2, nums)  # [1, 4, 9, 16, 25]

# Keep even squares
even_squares = filter(lambda x: x % 2 == 0, squares)  # [4, 16]

# Sum them
result = reduce(lambda x, y: x + y, even_squares)  # 20
print(result)

```python
#Write a Python function that checks if a given number is prime or not from 1 to 200
def is_prime(number):
    if number <= 1:
        return False
    if number <= 3:
        return True
    if number % 2 == 0 or number % 3 == 0:
        return False
```

```python
# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.current = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.max_terms:
            raise StopIteration
        if self.current == 0:
            self.current += 1
            return 0
        elif self.current == 1:
            self.current += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.current += 1
            return self.a
```

```python
#Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
```

```python
#Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
#Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, string))
```

```python
#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 € -
def calculate_order_values(orders):
    result = []
    for index, (price, quantity) in enumerate(orders, start=1):
        total = price * quantity
        if total < 100:
            total += 10
        result.append((index, total))
    return result

# Example usage:
orders = [(15.0, 2), (40.0, 1), (10.0, 10), (25.0, 5)]
order_values = calculate_order_values(orders)
print(order_values)
```

```python
#Write a Python program using lambda and map.
# Sample order list: (price per item, quantity)
orders = [(15.0, 2), (40.0, 1), (10.0, 10), (25.0, 5)]

# Map with lambda to compute (order_number, adjusted total price)
result = list(map(
    lambda x: (
        x[0] + 1,
        x[1][0] * x[1][1] + (10 if x[1][0] * x[1][1] < 100 else 0)
    ),
    enumerate(orders)
))

# Print result
print(result)
```
